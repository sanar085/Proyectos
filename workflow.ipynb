```python
def precsv(filename):
    with open(path+filename,encoding="Windows 1252") as infile:
        content = infile.read()
        soup = BeautifulSoup(content, 'html.parser')
        table = soup.select_one("table")
        return [[cell.text.replace("\n","") for cell in row.find_all(["td","th"])] for row in table.select("tr")]
```

En este caso seguimos con la codificación Widows 1252 y tenemos caracteres como "\u200b", "\xa0" y "\n", por lo que podemos probar a normalizar el texto a través de la libreria unicodedata. Para ello hay que usar el método normalize: unicodedata.normalize ('NFKD',cell.text).

```python
import unicodedata
```

```python
def precsv(filename):
    with open(path+filename,encoding="Windows 1252") as infile:
        content = infile.read()
        soup = BeautifulSoup(content, 'html.parser')
        table = soup.select_one("table")
        return [[unicodedata.normalize('NFKD',cell.text) for cell in row.find_all(["td","th"])] for row in table.select("tr")]
```

```python
precsv("idiomas.html")
```

Al ver este resultado, solo se me ocurre emplear el método .replace() para los caracteres que queremos eliminar. En el caso de "\u200b", se utiliza para indicar un espacio que no debe tener anchura visual, es decir, no se debe mostrar como un espacio en blanco. Su propósito principal es proporcionar una separación entre caracteres sin afectar la apariencia visual del texto, por lo que lo reemplazaré por la nada (""),al igual que "\n". En el caso de "\xa0", lo reemplazaré por un espacio.

```python
def precsv(filename):
    with open(path+filename,encoding="Windows 1252") as infile:
        content = infile.read()
        soup = BeautifulSoup(content, 'html.parser')
        table = soup.select_one("table")
        return [[cell.text.replace("\u200b","").replace("\xa0"," ").replace("\n","") for cell in row.find_all(["td","th"])] for row in table.select("tr")]
```

```python
precsv("idiomas.html")
```
